```python
from openai import OpenAI
from langchain_community.chat_message_histories import ChatMessageHistory

# Connect to LM Studio local server
client = OpenAI(
    base_url="http://127.0.0.1:1234/v1", 
    api_key="lm-studio"  # any string works
)
```

```python
history = ChatMessageHistory()

print("Chat started! Type 'exit' in the input box to stop.\n")

# Make sure this matches the 'Model ID' shown in your LM Studio Server tab
SELECTED_MODEL = "qwen3-1.7b" 

while True:
    query = input("User (Type 'exit' to quit): ")

    if query.lower() == 'exit':
        print("\nExiting...")
        print("Program Finished.")
        break
        
    if not query.strip():
        continue

    print(f"User Input: {query}")
    print('- ' * 50)
    
    # Add user message to history
    history.add_user_message(query)

    # Convert messages for the API
    messages_with_history = []
    for m in history.messages:
        role = "user" if m.type == "human" else "assistant"
        messages_with_history.append({"role": role, "content": m.content})

    try:
        # Request in LM Studio
        stream = client.chat.completions.create(
            model=SELECTED_MODEL,
            messages=messages_with_history,
            stream=True
        )
        
        full_ai_response = ""
        
        # Stream the response to the terminal
        for chunk in stream:
            if chunk.choices[0].delta.content:
                content = chunk.choices[0].delta.content
                print(content, end="", flush=True)
                full_ai_response += content

        # Save AI response to history   
        history.add_ai_message(full_ai_response)
        print('\n')
        print('=' * 100)

    except Exception as e:
        print(f"\nMay Error: {e}")
        print("Tip: Check if the 'Start Server' button is green in LM Studio.")
        break
```
