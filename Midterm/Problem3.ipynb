```python
import control
import matplotlib.pyplot as plt
```

```python
# System Definition
# G(s) = 3 / (8s^4 + 12s^3 + 26s^2 + 18s)
num = [3]
den = [8, 12, 26, 18, 0]
system = control.TransferFunction(num, den)

print("System Transfer Function G(s):")
print(system)
```

```python
# Stability Analysis
poles = control.poles(system)
zeros = control.zeros(system)

print("\n--- Stability Analysis ---")
print(f"Poles: {poles}")
print(f"Zeros: {zeros}")
```

```python
# Time-Domain Analysis: Step Response
plt.figure(figsize=(6, 4))
t, y = control.step_response(system)
plt.plot(t, y)
plt.title('Step Response')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
# Stability Margins
gm, pm, wg, wp = control.margin(system)
print("\n--- Frequency-Domain Stability Margins ---")
print(f"Gain Margin: {gm}")
print(f"Phase Margin: {pm} degrees")
print(f"Phase Crossover Frequency (for GM): {wg} rad/s")
print(f"Gain Crossover Frequency (for PM): {wp} rad/s")
```

```python
# Pole-Zero Map
plt.figure(figsize=(6, 4))
control.pzmap(system, grid=True, title='Pole-Zero Map') # Pass your 'system' object here

plt.show()
```
