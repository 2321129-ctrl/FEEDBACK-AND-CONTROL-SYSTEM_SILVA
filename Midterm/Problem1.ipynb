```python
import control as ctrl
import matplotlib.pyplot as plt
```

```python
# Transfer Function Construction
# G(s) = (s + 1) / (s^4 + 4s^3 + 7s^2 + 6s + 3)
num = [1, 1]
den = [1, 4, 7, 6, 3]
sys = control.TransferFunction(num, den)

```

```python
# Compute Poles and Zeros
poles = control.poles(sys)
zeros = control.zeros(sys)
```

```python
# Calculate Stability Margins
gm, pm, wg, wp = control.margin(sys)
```

```python
# Display Poles, Zeros, and Stability Margins
print("--- System Analysis Results ---")
print(f"System: {sys}")
print(f"\nPoles of the system:\n{poles}")
print(f"\nZeros of the system:\n{zeros}")
print(f"\nStability Margins:")
print(f"Gain Margin: {gm:.4f} (at {wg:.4f} rad/s)")
print(f"Phase Margin: {pm:.4f} degrees (at {wp:.4f} rad/s)")
print("-" * 30)
```

```python
# Performance Metrics
# Calculate step response and metrics
t, y = ct.step_response(sys)
info = ct.step_info(sys)
```
